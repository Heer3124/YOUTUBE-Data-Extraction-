```python
!pip install selenium webdriver-manager
!apt-get update
!wget https://dl.google.com/linux/direct/google-chrome-stable_current_amd64.deb
!dpkg -i google-chrome-stable_current_amd64.deb
!apt-get install -f # Install dependencies if any

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 1. Setup Chrome options for headless execution in Colab
options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--headless') # Re-added for headless execution
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-gpu') # Added to potentially improve stability
options.add_argument('--window-size=1920x1080') # Set a default window size
options.add_argument('--disable-notifications') # Blocks pop-ups
options.add_argument('--disable-extensions') # Disable browser extensions
options.binary_location = '/usr/bin/google-chrome-stable' # Specify the path to Google Chrome Stable

# 2. Open the Browser using webdriver_manager
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

try:
    # 3. Go to the YouTube Channel
    url = "https://www.youtube.com/@SRPAY/videos"
    print(f"Opening: {url}")
    driver.get(url)

    # Attempt to handle cookie consent if it appears
    try:
        # Wait for and click the "Accept all" button for cookies if it appears
        accept_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(., 'Accept all')]" ))
        )
        accept_button.click()
        print("Clicked 'Accept all' on cookie consent dialog.")
        time.sleep(2) # Give it a moment to process the click and for the page to refresh
    except Exception:
        print("No 'Accept all' consent button found or could not click it. Continuing...")
        pass # Continue if no consent dialog is present or button not found

    # 4. Use an explicit wait to ensure the video elements are loaded
    # We wait up to 20 seconds for at least one video title link to be present
    wait = WebDriverWait(driver, 20)
    wait.until(EC.presence_of_element_located((By.ID, "video-title-link")))

    # 5. Hand the "Rendered" HTML over to BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # 6. Find the video titles
    video_tags = soup.find_all('a', id='video-title-link')

    print(f"\n--- Found {len(video_tags)} Videos ---\n")

    for index, tag in enumerate(video_tags, 1):
        title = tag.get_text().strip()
        link = "https://www.youtube.com" + tag.get('href')
        print(f"{index}. {title}")
        print(f"   URL: {link}\n")

finally:
    # 7. Close the browser window automatically
    driver.quit()
```
